# Identify the type of data in each column (categorical or numerical)?

Seperate out the categorical columns from the numerical types



```python
def separate(credit_data):
    categorical = []
    continuous = []
    for column in credit_data.columns:
        if credit_data[column].nunique() < 40:
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical, continuous

categorical, continuous = separate(credit_data)

from tabulate import tabulate

print(tabulate({"Categorical": categorical, "Continuous": continuous}, headers=["Categorical", "Continuous"]))

        
```

In this initial phase, EDA focuses on understanding the dataset by examining its structure, dimensions, and basic statistics. It involves tasks such as data cleaning, checking for missing values, and summarizing key statistical measures like mean, median, and standard deviation.

## Pattern and Relationship Discovery: 
EDA aims to uncover patterns, trends, and relationships within the data. This includes creating visualizations such as scatter plots, histograms, and box plots to identify distributions and correlations between variables. ## Hypothesis Testing and Insights:

# 1)What is the distribution of loan approval based on gender?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of loan approval based on gender
plt.figure(figsize=(8, 4))
ax = sns.countplot(data=credit_data, x='GENDER', hue='TARGET')
plt.title(" Gender Distribution and Loan Approval")
plt.xlabel("Gender")
plt.ylabel("Loan Approval Count")
plt.legend(title='Loan Approval', labels=['Not Default', 'Default'])

# Annotate the bars with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.show()
```

# The majority of loan applicants are female.


# 2) Do clients with a car have a different loan repayment behavior?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loan approval for clients with a car
plt.figure(figsize=(6, 4))
ax = sns.barplot(data=credit_data, x='Car', y='TARGET', ci=None)
plt.title(" Loan Approval for Clients with a Car")
plt.xlabel("Car Ownership")
plt.ylabel("Loan Approval Rate (0 - No, 1 - Yes)")
plt.ylim(0, 1)

# Annotate the bars with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.show()
```

# 16)How does the education level of clients impact loan approval and repayment?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming credit_data contains your data
# Aggregate the data by the 'NAME_EDUCATION_TYPE' column and calculate the mean of 'TARGET'
agg_data = credit_data.groupby('NAME_EDUCATION_TYPE')['TARGET'].mean().reset_index()

# Create a bar plot for the aggregated data
plt.figure(figsize=(12, 6))
sns.barplot(data=agg_data, x='NAME_EDUCATION_TYPE', y='TARGET', ci=None)
plt.title("Impact of Education Level on Loan Approval and Repayment (Aggregated Data)")
plt.xlabel("Education Level")
plt.ylabel("Average Loan Approval Rate (0 - No, 1 - Yes)")

# Add counts on top of the bars
for i, row in agg_data.iterrows():
    plt.text(i, row['TARGET'], f"{row['TARGET']:.2f}", ha='center', va='bottom')

plt.xticks(rotation=45)
plt.show()

```

# Clients with higher education levels tend to have a higher probability of loan approval and repayment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming credit_data contains your data

# Convert 'DAYS_EMPLOYED' to years of employment (assuming 365 days in a year)
credit_data['YEARS_EMPLOYED'] = -credit_data['DAYS_EMPLOYED'] / 365

# Define bins for years of employment
bins = [0, 1, 3, 5, 10, 20, 30, 40, 50]  # Customize the bins as needed

# Create labels for the bins (e.g., '0-1', '1-3', '3-5', etc.)
labels = [f"{start}-{end}" for start, end in zip(bins[:-1], bins[1:])]

# Assign each row to a bin based on 'YEARS_EMPLOYED'
credit_data['EMPLOYMENT_BIN'] = pd.cut(credit_data['YEARS_EMPLOYED'], bins=bins, labels=labels)

# Aggregate the data by 'EMPLOYMENT_BIN' and calculate the mean of 'TARGET'
agg_data = credit_data.groupby('EMPLOYMENT_BIN')['TARGET'].mean().reset_index()

# Create a bar plot for the aggregated data
plt.figure(figsize=(10, 6))
sns.barplot(data=agg_data, x='EMPLOYMENT_BIN', y='TARGET', ci=None)
plt.title("Impact of Length of Employment on Loan Repayment Behavior (Aggregated Data)")
plt.xlabel("Years of Employment")
plt.ylabel("Average Loan Repayment Rate (0 - Default, 1 - Repaid)")

# Add counts on top of the bars
for i, row in agg_data.iterrows():
    plt.text(i, row['TARGET'], f"{row['TARGET']:.2f}", ha='center', va='bottom')

plt.xticks(rotation=45)
plt.show()

```
